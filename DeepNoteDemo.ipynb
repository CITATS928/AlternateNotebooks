```
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the survival rate by Pclass
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df_train, ci=None)
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

> Visualize the distribution of the target variable to understand its class balance

```
import matplotlib.pyplot as plt

# Visualize the distribution of the target variable 'Survived'
df_train['Survived'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribution of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'], rotation=0)
plt.show()
```

The bar chart above shows the distribution of the target variable 'Survived'. This helps us understand the class balance between those who survived and those who did not.

> Use a box plot to see the survival rate of passengers of different ages, and tell me what you find

```
plt.figure(figsize=(12, 6))
sns.boxplot(x='Survived', y='Age', data=df_train)
plt.title('Survival Rate by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

The box plot shows the distribution of ages for passengers who survived and those who did not. Here are some observations:

1. The median age of survivors is lower than that of non-survivors.
2. There is a wider age range among non-survivors compared to survivors.
3. Younger passengers (children) have a higher survival rate, as indicated by the lower median age and the presence of younger ages in the survivor group.

This suggests that younger passengers had a better chance of survival.

> Use heatmap to analyze the survival rate of cabin class and gender combination and tell me the conclusion

```
# Creating a pivot table to analyze the survival rate by Pclass and Sex
pivot_table = df_train.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean')

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', cbar=True)
plt.title('Survival Rate by Cabin Class and Gender')
plt.xlabel('Gender')
plt.ylabel('Pclass')
plt.show()
```

The heatmap shows the survival rates for different combinations of cabin class (Pclass) and gender (Sex). Here are the conclusions:

1. Females had a higher survival rate across all cabin classes compared to males.
2. The highest survival rate was observed for females in the first class (Pclass 1).
3. Males in the third class (Pclass 3) had the lowest survival rate.
4. The survival rate for males increased with higher cabin classes, but it was still lower than that of females in the same class.

Overall, gender and cabin class both played significant roles in determining the survival rate, with females and higher cabin classes having better chances of survival.
